```python
import math
import matplotlib.pyplot as plt

def exponential_decay(initial_weight, final_weight, num_epochs):
    decay_rate = -math.log(final_weight / initial_weight) / num_epochs
    weights = [initial_weight * math.exp(-decay_rate * epoch) for epoch in range(num_epochs)]
    return weights

# 初期重み
initial_weight = 7
final_weight = 0.1
# エポック数
num_epochs = 30

# 指数関数的に減衰する重みのリストを取得
weights = exponential_decay(initial_weight, final_weight, num_epochs)

print(len(weights))
# 結果を表示
print("Epoch\tWeight")
#for epoch, weight in enumerate(weights):
    #print(f"{epoch+1}\t{weight:.4f}")
plt.plot(weights)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# 256次元の値をランダムに生成（実際のデータに置き換える必要があります）
data = np.random.rand(256)

# 16x16の行列に変換
matrix = data.reshape((16, 16))

# Figureオブジェクトを作成します
fig = plt.figure()

# 画像を描画します
plt.imshow(matrix, cmap='hot', interpolation='nearest')
plt.colorbar()  # カラーバーの表示

# 画像をファイルに保存します
fig.savefig('heatmap.png')

# 画像を表示します（plt.show()をコメントアウトして画像のみを保存）
#plt.show()
```
